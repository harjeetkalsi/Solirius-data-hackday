```python
import geopandas as gpd
uk_map = gpd.read_file('/home/sol-022/hackday/Solirius-data-hackday/Regions__December_2017__Boundaries.shp')
```

```python
dataset_covid_persons = dataset_covid_only[dataset_covid_only["Sex"] == 'Persons']
```

```python
merged = uk_map.set_index("rgn17nm").join(dataset_covid_persons.set_index("Area of usual residence name"))
```

```python
merged
```

```python
merged.plot()
```

```python
# set a variable that will call whatever column we want to visualise on the map
deaths = "Deaths"
# set the range for the choropleth
vmin, vmax = 0, merged["Deaths"].max()
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 6))
# create map
merged.plot(column=deaths, cmap="Blues", linewidth=0.8, ax=ax, edgecolor="0.8")
ax.axis("off")
ax.set_title("Covid deaths in England between March - May 2020", fontdict={"fontsize": "25", "fontweight" : "3"})

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap="Blues", norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
```

```python
from bokeh.io import curdoc, output_notebook
from bokeh.models import Slider, HoverTool
from bokeh.layouts import widgetbox, row, column
import json

# add region as a column
merged["Region"] = merged.index

#Read data to json.
merged_json = json.loads(merged.to_json())
#Convert to String like object.
json_data = json.dumps(merged_json)
```

```python
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer
from bokeh.plotting import figure
from bokeh.io import output_notebook, show, output_file
```

```python
from bokeh.io import output_notebook, show, output_file, push_notebook
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LinearColorMapper, ColorBar
from bokeh.palettes import brewer
#Input GeoJSON source that contains features for plotting.
geosource = GeoJSONDataSource(geojson = json_data)
#Define a sequential multi-hue color palette.
palette = brewer['YlGnBu'][8]
#Reverse color order so that dark blue is highest obesity.
palette = palette[::-1]
#Instantiate LinearColorMapper that linearly maps numbers in a range, into a sequence of colors.
color_mapper = LinearColorMapper(palette = palette, low = 0, high =  merged["Deaths"].max())
#Add hover tool
hover = HoverTool(tooltips = [("Region", "@Region"), ('Deaths', '@Deaths')])

#Define custom tick labels for color bar.
tick_labels = {'0': '0%', '5': '5%', '10':'10%', '15':'15%', '20':'20%', '25':'25%', '30':'30%','35':'35%', '40': '>40%'}
#Create color bar. 
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 500, height = 20,
border_line_color=None,location = (0,0), orientation = 'horizontal', major_label_overrides = tick_labels)
#Create figure object.
p = figure(title = 'England Covid deaths March - May 2020', plot_height = 600 , plot_width = 600, toolbar_location = None, tools=[hover])
p.axis.visible = False
p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None
#Add patch renderer to figure. 
p.patches('xs','ys', source = geosource,fill_color = {'field' :'Deaths', 'transform' : color_mapper},
          line_color = 'black', line_width = 0.25, fill_alpha = 1)
#Specify figure layout.
p.add_layout(color_bar, 'below')
#Display figure inline in Jupyter Notebook.
output_notebook()
#Display figure.
show(p, notebook_handle=True)
push_notebook()
```
