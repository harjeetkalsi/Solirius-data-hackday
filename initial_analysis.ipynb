```python
import seaborn as sns
```

```python
dataset_covid_only = dataset[dataset["Cause of death"] == 'COVID-19']
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
ax = sns.barplot(x="Deaths", y="Area of usual residence name", data=dataset_covid_only, ci = None).set_title('Covid deaths by region in UK between March - May 2020', fontsize=30, pad=30)
```

```python
sns.barplot(x='Sex', y='Deaths',data=dataset_covid_only, ci = None).set_title('Covid deaths in UK between March - May 2020', fontsize=30, pad=30)
```

```python
dataset_without_person = dataset_covid_only[dataset_covid_only["Sex"] != 'Persons']
```

```python
dataset_without_person
```

```python
colors = ["#fc6b03", "#fc9d03", "#fc0324"]
sns.set_palette(colors)
sns.barplot(x="Deaths", y="Area of usual residence name", data=dataset_without_person, hue = dataset['Sex'], ci = None).set_title("Covid deaths by region and gender in UK between March - May 2020", fontsize=30, pad=30)
```

```python
import geopandas as gpd
uk_map = gpd.read_file('/home/sol-022/hackday/Solirius-data-hackday/Regions__December_2017__Boundaries.shp')
```

```python
dataset_covid_persons = dataset_covid_only[dataset_covid_only["Sex"] == 'Persons']
```

```python
merged = uk_map.set_index("rgn17nm").join(dataset_covid_persons.set_index("Area of usual residence name"))
```

```python
merged
```

```python
merged.plot()
```

```python
# set a variable that will call whatever column we want to visualise on the map
deaths = "Deaths"
# set the range for the choropleth
vmin, vmax = 0, merged["Deaths"].max()
# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 6))
# create map
merged.plot(column=deaths, cmap="Blues", linewidth=0.8, ax=ax, edgecolor="0.8")
ax.axis("off")
ax.set_title("Covid deaths in England between March - May 2020", fontdict={"fontsize": "25", "fontweight" : "3"})

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap="Blues", norm=plt.Normalize(vmin=vmin, vmax=vmax))
# empty array for the data range
sm._A = []
# add the colorbar to the figure
cbar = fig.colorbar(sm)
```
